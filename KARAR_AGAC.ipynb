```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True)
df
```

```python
M=df[df.diagnosis=="M"]
B=df[df.diagnosis=="B"]
```

```python
B
```

```python
B = B.head(20)
B
```

```python
M
```

```python
df = pd.concat([M, B])
df
```

```python
df["diagnosis"].value_counts()
```

```python
data=df
df.diagnosis=[0 if each == "M" else 1 for each in data.diagnosis]
df
```

```python
y=df.diagnosis.values
y
```

```python
import numpy as np
```

```python
x_data=df.drop(["diagnosis"],axis=1)
x_data
```

```python
x=(x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
tahmin=knn.predict(x_test)
tahmin
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,tahmin))
```

```python
from sklearn.metrics import confusion_matrix

# y_pred ve y_test verileri mevcut olsun
cm = confusion_matrix(y_test, tahmin)
print(cm)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
precision = precision_score(y_test, tahmin)

# Recall hesaplama
recall = recall_score(y_test, tahmin)

print("Precision: ", precision)
print("Recall: ", recall)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, tahmin, average='weighted')
print("F1 score:", f1)
```
